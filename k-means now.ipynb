```python
from sklearn.cluster import KMeans
from sklearn import preprocessing
import sklearn.cluster as cluster
import sklearn.metrics as metrics
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python

sns.set()
from sklearn.cluster import KMeans
##Loading the Data
df = pd.read_csv('mac1.csv')
df
```

```python
df["Degree"].value_counts()
```

```python
df["label"] = (df["Degree"] >= 21.5 ).astype(int)
```

```python
df.head(5)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
```

```python
pipeline = Pipeline([("kmeans", KMeans(n_clusters = 4)), ("dec", DecisionTreeClassifier())])
```

```python
x, y = df[["AverageShortestPathLength"]], df.iloc[:, 3]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 10, train_size = 0.8)
```

```python
pipeline.fit(x_train, y_train)
```

```python
pipeline.score(x_test, y_test)

pipeline["kmeans"].cluster_centers_

y_pred = pipeline.predict(x_test)
x_test_new = x_test.copy()
x_test_new["quality"] = y_test
x_test_new["predicted quality"] = y_pred
print(x_test_new.head(20))

print(y_pred)
```

```python

from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score

precision_score(y_test, y_pred)
```

```python

recall_score(y_test, y_pred)
```

```python

print(f1_score(y_test, y_pred))
```
